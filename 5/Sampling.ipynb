```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
style.use('ggplot')
t = np.arange(-1, 1, 0.00001)
```

```python
def x1(t):
    return np.cos(10 * np.pi * t)
```

# 1

```python
plt.plot(t, x1(t), label="cos(10pit)")
plt.legend()
plt.show()
```

الف) تابع نمونه‌برداری

```python
def sampling(input_signal,sampling_freq):
    time_s = 1 / sampling_freq
    ratio_s = int(time_s // 0.00001)
    sampled_signal = [input_signal[i * ratio_s] for i in range(int(len(input_signal) // ratio_s)+1)]
    sampled_time =   [t[i * ratio_s] for i in range(int(len(t)//ratio_s)+1)]
    return sampled_signal,sampled_time
    
```

ب) با فرکانس ۴ هرتز نمایش ‌داده شد

```python
plt.plot(t, x1(t), label="cos(10pit)")
sampled_signal , sampled_t = sampling(x1(t),4)
plt.plot(sampled_t, sampled_signal, label='4Hz')
plt.legend()
plt.show()
```

ج) خیر، مشاهده می‌کنیم بسیاری از نقاط نادیده گرفته شده‌اند
