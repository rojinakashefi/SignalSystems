```python
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from matplotlib import style
import numpy as np
```

```python
style.use('ggplot')
t = np.arange(-15, 15, 0.01)
```

```python
def fourier_transform(t, x, w):
    return (x(t) * np.exp(-1j * w * t)).real
```

```python
def signal_ft (x):
    return [integrate.quad(fourier_transform, -15, 15, args=(x, w)) for w in t]
```

```python
def plot_s(x):
    figure, axis = plt.subplots(2)
    axis[0].plot(t, x(t))
    axis[1].plot(t, signal_ft(x))
    plt.show()
```

```python
def A(t):
    return - np.heaviside(t-1, 1) - np.heaviside(-t - 1, 1) + 1
```

```python
plot_s(A)
```

```python
def B(t):
    return (1 / (np.pi)) * np.sinc(t / np.pi)
```

```python
plot_s(B)
```

```python
def C(t):
    return np.cos(t) * ((3 / (np.pi)) * np.sinc((3*t)/np.pi))
```

```python
plot_s(C)
```

```python
def D(t):
    return B(t) ** 2
```

```python
plot_s(D)
```
